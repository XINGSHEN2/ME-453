```python
# Package import
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
# from matplotlib.ticker import(AutoMinorLocator, MultipleLocator)

# Plot parameter setting 
plt.style.use('default')
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['figure.figsize'] = 8, 6
plt.rcParams['figure.dpi'] = 100

```

The following code generates the control chart in Topic 5 P.36. Once you understand how it works, you should be able to generate your own control chart.

```python
# Control chart example
sample = pd.read_csv('sample.csv', header = None, index_col = None, names=["1", "2", "3", "4", "5", "xbar", "R"])
xbb = np.mean(sample.xbar)
rb = np.mean(sample.R)

A2 = 0.577
xUCL = xbb + A2*rb
xCL = xbb
xLCL = xbb - A2*rb
D3 = 0
D4 = 2.114
sUCL = D4*rb
sCL = rb
sLCL = D3*rb

# xb chart
ax1 = plt.subplot(211)
ax1.plot(sample["xbar"],'c', marker='o', markersize='5')
plt.plot([-1, 26], [xUCL, xUCL], 'b--')
plt.plot([-1, 26], [xCL, xCL], 'b--')
plt.plot([-1, 26], [xLCL, xLCL], 'b--')
plt.text(26.5, xUCL, 'UCL='+str(round(xUCL, 4)), color='b', fontsize=12)
plt.text(26.5, xCL, 'CL='+str(round(xCL, 4)), color='b', fontsize=12)
plt.text(26.5, xLCL, 'LCL='+str(round(xLCL, 4)), color='b', fontsize=12)
plt.ylabel('Sample mean')
plt.xlim(-1, 26)
plt.ylim(xLCL-0.05, xUCL+0.05)
plt.grid(axis='y')
plt.tight_layout()

# R chart
ax2 = plt.subplot(212)
ax2.plot(sample["R"], 'c', marker='o', markersize='5')
plt.plot([-1, 26], [sUCL, sUCL], 'b--')
plt.plot([-1, 26], [sCL, sCL], 'b--')
plt.plot([-1, 26], [sLCL, sLCL], 'b--')
plt.text(26.5, sUCL, 'UCL='+str(round(sUCL, 4)), color='b', fontsize=12)
plt.text(26.5, sCL, 'CL='+str(round(sCL, 4)), color='b', fontsize=12)
plt.text(26.5, sLCL, 'LCL='+str(round(sLCL, 4)), color='b', fontsize=12)
plt.ylabel('Sample range')
plt.xlabel('Sample index')
plt.xlim(-1, 26)
plt.ylim(sLCL-0.05, sUCL+0.05)
plt.grid(axis='y')
plt.tight_layout()

plt.subplots_adjust(top=0.93)
plt.suptitle(r'$\bar{X}-R$ chart', fontsize=14)

print("xbb = ", xbb)
print("rb = ", rb)
```
