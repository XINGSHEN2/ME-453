```python
# problem 4.8
from scipy.stats import ttest_ind

configuration_A = welding_df[welding_df['Configuration'] == 'A']
configuration_B = welding_df[welding_df['Configuration'] == 'B']
configuration_C = welding_df[welding_df['Configuration'] == 'C']

for i in welding_df.columns[1:]:
    
    print(i)
    result_AB = ttest_ind(configuration_A[i], configuration_B[i], equal_var=False)
    result_AC = ttest_ind(configuration_A[i], configuration_C[i], equal_var=False)

    alpha = 0.05  # significance level

    if result_AB.pvalue < alpha:
        print("Reject the null hypothesis: Configurations A and B have different mean values.")
    else:
        print("Fail to reject the null hypothesis: Configurations A and B have the same mean values.")

    if result_AC.pvalue < alpha:
        print("Reject the null hypothesis: Configurations A and C have different mean values.")
    else:
        print("Fail to reject the null hypothesis: Configurations A and C have the same mean values.")
        
    print("---------------")
```
