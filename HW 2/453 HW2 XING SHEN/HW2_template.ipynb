```python
# Package import
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
# Problem 1.1
np.random.seed(99)
N =  100 # 100 set of samples
n =  10 # sample size
sigma = np.sqrt(0.01)
mu = 10
sample = sigma*np.random.randn(N, n) + mu
mean = np.zeros((N,)) 
cover = np.zeros((N,)) 
alpha = 0.05
CI = st.norm.ppf(1 - alpha / 2) * sigma / np.sqrt(n) # Confidence interval, known variance
for i in range(100):
    mean[i] = np.mean(sample[i,:])# sample mean
    cover[i] = 1 if mean[i] - CI <= mu <= mean[i] + CI else -1 # Check whether CI cover the true mean

# Note: No plot needed for Problem 1.1
```

```python
# Problem 1.2
Bad = np.where(cover == -1) # All not covered sample
# Good: Find all "covered" sample
Good = np.where(cover == 1)
plt.plot([0, 100], [mu, mu], '--', color='orange', linewidth=1, label='true mean') 
plt.errorbar(Bad[0], mean[Bad[0]], yerr = CI, marker='o', ms=3, color='r', ls='', ecolor='r', elinewidth=0.5, capsize=2, label='Fail') # Error bar for Bad
# Try to play with the parameters above and see what changes!

# Your code for Error bar of Good
plt.errorbar(Good[0], mean[Good[0]], yerr = CI, marker='o', ms=3, color='b', ls='', ecolor='b', elinewidth=0.5, capsize=2, label='Success')

# Your code for necessary plot information (legned, xy axis, etc.)
plt.xlabel('Sample Index')
plt.ylabel('Sample Mean')
plt.title('CI for Sample Means')
plt.legend()
plt.show()
```

```python
# Problem 2.1-3
sigma = 2
mu = 16
alpha = 0.05
delta = 16 - 20 # mean shift
sample_n = np.arange(1,11,1)
Z_alpha2 = st.norm.ppf(1- alpha/2)
beta = np.zeros((10,))

for i in sample_n: 
    beta[i-1] = st.norm.cdf(Z_alpha2 - (delta * np.sqrt(i)) / sigma) - st.norm.cdf(-Z_alpha2 - (delta * np.sqrt(i)) / sigma) # Your code to caculate beta for given i
    print("n = ", i, ", beta = ", beta[i-1])
# Use a plot to show the relation between sample_n and beta
plt.plot(sample_n, beta, marker='o', linestyle='-', color='b')
plt.xlabel('Sample Size (n)')
plt.ylabel('Probability of Type II Error (beta)')
plt.title('Power Analysis')
plt.grid(True)
plt.show()
```

```python
# Problem 3.3
n = 4
alpha = 0.0001
shift = np.arange(0,5,0.1)
Z_alpha2 = st.norm.ppf(1- alpha/2)
beta = np.zeros((50,))

for i in range(50):
    beta[i] = st.norm.cdf(Z_alpha2 - shift[i]*np.sqrt(n)) - st.norm.cdf(-Z_alpha2 - shift[i]*np.sqrt(n)) # Your code to caculate beta for given i
    print("n = ", i, ", beta = ", beta[i])
# Use a plot to show the relation between sample_n and beta
plt.plot(shift, beta, marker='o', linestyle='-', color='b')
plt.xlabel('shift')
plt.ylabel('Probability of Type II Error (beta)')
plt.title('Power Analysis')
plt.grid(True)
plt.show()
```
