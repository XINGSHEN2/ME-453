```python
# Class define
class Data:
    def __init__(self) -> None:
        # Parameter list
        self.location = "./Data/"
        self.parameter = pd.read_excel("Parameter.xlsx", header=0)
        self.parameter.sort_values(["Length", "Surface"], inplace=True, ignore_index=True)
        self.parameter.index += 1
    
    #columns: 'acoustic', 'LVDT', 'power', 'microphone', 'pressure'
    def readPower(self, cycle):
        path = self.location + str(cycle) + ".csv"
        df = pd.read_csv(path, usecols=[2], names=["data"])
        return df

    def readAE(self, cycle):
        path = self.location + str(cycle) + ".csv"
        df = pd.read_csv(path, usecols=[0], names=["data"])
        return df

    def readMic(self, cycle):
        path = self.location + str(cycle) + ".csv"
        df = pd.read_csv(path, usecols=[3], names=["data"])
        return df
```

"welding.readAE()" is a function to get the AE signal from specific cycle. There is also a "welding.readMic" to get the microphone signal.
The following code show how to do fft in Python, and the figure give you more idea on how it works. Note that I did inverse fft for demo purpose only. If you did fft correctly, you can retrieve the signal and it should be identical to the raw signal. Once you see how the code works, it should be straight forward to get all peak frequency and power.

```python
# Problem 3
signal = welding.readAE(13991)["data"] # signal is a dataframe
signal = signal - np.mean(signal)
Fs = 200e3 # Sampling rate
T = 2 # Duration
N = int(T*Fs)
yf = np.fft.fft(signal)
yif = np.fft.ifft(yf) # Inverse fft (For demo purpose)
yf = yf/N # Normalize intensity
yf = yf[:N//2] # One side is enough since the fft is symmetric

xf = np.fft.fftfreq(N, 1/Fs)
xf = xf[:N//2]

plt.figure(figsize=(16, 8), dpi=72)
plt.subplot(211)
plt.plot(xf, abs(yf), "b-")
plt.grid(axis="x")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Intensity')

plt.subplot(212)
plt.plot(np.linspace(0, T, N), signal, "b")
plt.grid()
plt.xlabel("Time (ms)")
plt.ylabel("AE (V)")

index_20kHz = int(20e3*T)
peak_freq_power = np.max(abs(yf)[(index_20kHz - 1000):(index_20kHz + 1000)])
peak_freq = xf[np.where(abs(yf) == peak_freq_power)]
print('The peak frequency around 20 kHz is %.2f Hz and the corresponding power value is %.4f' % (peak_freq, peak_freq_power))

# Hint: You have peak frequency and power for specific cycle. How do you get these information for all cycle which is "New Tool & Clean Surface"? 


# Use inverse fft to retrieve the original signal (It should be same as the original signal!!)
plt.figure(figsize=(16, 4), dpi=72)
plt.plot(np.linspace(0, T, N), yif.real, linestyle='-', color='blue')
plt.grid()
plt.xlabel('t (s)')
plt.ylabel('amplitude')
plt.title("Inverse FFT")
```

```python
#%% FFT_AE (@20 kHz) result
PM["FFT_Mic_f"] = ""
PM["FFT_Mic_P"] = ""
PM["FFT_AE_f"] = ""
PM["FFT_AE_P"] = ""
index_20kHz = int(20e3*T)
xf = np.fft.fftfreq(N, 1/Fs)
xf = xf[:N//2]

for cycle in PM.Cycle:
    # Mic
    signal = welding.readMic(cycle)["data"]
    signal = signal - np.mean(signal)
    yf = np.fft.fft(signal)
    yf = yf/N # Normalize intensity
    yf = yf[:N//2] # One side is enough since the fft is symmetric
    peak_freq_power = max(abs(yf)[(index_20kHz - 1000):(index_20kHz + 1000)])
    peak_freq = xf[np.where(abs(yf) == peak_freq_power)]
    PM.loc[PM.Cycle == cycle, "FFT_Mic_P"] = peak_freq_power
    PM.loc[PM.Cycle == cycle, "FFT_Mic_f"] = peak_freq

    # AE
    signal = welding.readAE(cycle)["data"]
    signal = signal - np.mean(signal)
    yf = np.fft.fft(signal)
    yf = yf/N # Normalize intensity
    yf = yf[:N//2] # One side is enough since the fft is symmetric
    peak_freq_power = max(abs(yf)[(index_20kHz - 1000):(index_20kHz + 1000)])
    peak_freq = xf[np.where(abs(yf) == peak_freq_power)]
    PM.loc[PM.Cycle == cycle, "FFT_AE_P"] = peak_freq_power
    PM.loc[PM.Cycle == cycle, "FFT_AE_f"] = peak_freq

# PM.to_csv(".\FFT_result.csv")
```

#### Paired t-test
$$H_0: \mu_1 = \mu_2 \quad H_1: \mu_1 \neq \mu_2$$
$$
t = \frac{\bar{X}_1 - \bar{X}_2}{s_p\sqrt{\frac{2}{n}}}
\quad\quad s_p = \sqrt{\frac{s_1^2 + s_2^2}{2}}
\quad\quad df = 2n-2
$$
If $|t| > t_{(1-\alpha/2), df}$, reject $H_0$.

```python
# Pair t-test

# t-test funciton
def Pair_test(xb1, xb2, s1, s2):
    t_critical = stat.t.ppf(1-0.05/2, 58)
    sp = np.sqrt((s1**2 + s2**2)/2)
    t = (xb1-xb2)/sp/np.sqrt(2/30)
    if abs(t) > t_critical: 
        print(abs(t), t_critical, " - Reject")
        return
    print(abs(t), t_critical, " - Not Reject")

# FFT = pd.read_csv(".\FFT_result.csv")
target = "FFT_Mic_P" # Target signal
length_list = [0, 50, 60, 70, 80]
mean = np.zeros((5,))
std = np.zeros((5,))

for i in range(1,5):
    data = FFT[(FFT.Length == length_list[i]) & (FFT.Surface == "Clean")]
    mean[i] = np.mean(data[target])
    std[i] = np.std(data[target], ddof=1)
    
Pair_test(mean[1], mean[2], std[1], std[2])
Pair_test(mean[1], mean[3], std[1], std[3])
Pair_test(mean[1], mean[4], std[1], std[4])
Pair_test(mean[2], mean[3], std[2], std[3])
Pair_test(mean[2], mean[4], std[2], std[4])
Pair_test(mean[3], mean[4], std[3], std[4])
```
